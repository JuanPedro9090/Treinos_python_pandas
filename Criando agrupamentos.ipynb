# Formas de calcular a média com base em uma coluna

```python
grupo_bairro = dados1.groupby('Bairro')
```

```python
type(grupo_bairro)
```

```python
grupo_bairro.groups
```

```python
for bairro, data in grupo_bairro:
    print('{} -> {}'.format(bairro, data['Valor'].mean()))
```

```python
grupo_bairro['Valor', 'Condominio'].mean().round(2)
```

# Estatísticas Descritivas

```python
grupo_bairro['Valor'].aggregate(['min','max','sum','mean']).round(2).rename(columns={'min':'Minímo','max':'Máximo','sum':'Soma','mean':'Média'})
```
